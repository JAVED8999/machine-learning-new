```python
salary = df["salary"].value_counts(normalize=True)
salary = pd.DataFrame(salary)
salary = salary.reset_index()
salary = salary.rename(columns={"index":"salaryrange", "salary":"% of employee"})
salary = salary.sort_values(by='salaryrange')
salary["% of employee"]= salary["% of employee"]*100
salary["% of employee"]= np.round(salary["% of employee"],2)
display(salary)
```

```python
number_project = df["number_project"].value_counts(normalize=True)
number_project = pd.DataFrame(number_project)
number_project = number_project.reset_index()
number_project =number_project.rename(columns={"index":"project range", "number_project":"% of employee"})
number_project = number_project.sort_values(by='project range')
number_project["% of employee"]=number_project["% of employee"]*100
number_project["% of employee"]= np.round(number_project["% of employee"],2)
display(number_project)
```

```python
promotion_last_5years = df["promotion_last_5years"].value_counts(normalize=True)
promotion_last_5years = pd.DataFrame(promotion_last_5years)
promotion_last_5years = promotion_last_5years.reset_index()
promotion_last_5years = promotion_last_5years.rename(columns={"index":"promotion range", "promotion_last_5years":"% of employee"})
promotion_last_5years = promotion_last_5years.sort_values(by='promotion range')
promotion_last_5years["% of employee"]= promotion_last_5years["% of employee"]*100
promotion_last_5years["% of employee"]= np.round(promotion_last_5years["% of employee"],2)
display(promotion_last_5years)
```

```python
satisfaction_level = df["satisfaction_level"].value_counts(normalize=True)
satisfaction_level = pd.DataFrame(satisfaction_level)
satisfaction_level = satisfaction_level.reset_index()
satisfaction_level = satisfaction_level.rename(columns={"index":"satisfactionrange", "satisfaction_level":"% of employee"})
satisfaction_level = satisfaction_level.sort_values(by='satisfactionrange')
satisfaction_level["% of employee"]= satisfaction_level["% of employee"]*100
satisfaction_level["% of employee"]= np.round(satisfaction_level["% of employee"],2)
display(satisfaction_level)
```

```python
last_evaluation= df["last_evaluation"].value_counts(normalize=True)
last_evaluation = pd.DataFrame(last_evaluation)
last_evaluation = last_evaluation.reset_index()
last_evaluation = last_evaluation.rename(columns={"index":"last_evaluationrange", "last_evaluation":"% of employee"})
last_evaluation = last_evaluation.sort_values(by='last_evaluationrange')
last_evaluation["% of employee"]= last_evaluation["% of employee"]*100
last_evaluation["% of employee"]= np.round(last_evaluation["% of employee"],2)
display(last_evaluation)
```
